```python
import numpy as np
import matplotlib.pyplot as plt

L = 100
T = 10

h = 1
tau = 1
Nx = int(L/h) + 1
Nt = int(T/tau) + 1
x = np.linspace(0, L, num=Nx)   #равномерная сетка для пространственной координаты
u = np.zeros((Nt, Nx))     #решение    

#задание НУ 
u[0, :] = np.sin(8*np.pi*x/L)
     
for n in range(1,Nt):
    u_n = u[n-1].copy()
    F = np.zeros_like(u_n)
    for i in range(1, Nx-1):
        F[i] = (u_n[i]**2 - u_n[i-1]**2)/2
    
    for i in range(1, Nx-1):
        u[n,i] = u_n[i] - (F[i] - F[i-1])*(tau/h)
        



plt.figure(figsize=(8,6))
for i in range(0, Nt, 5 ):
    plt.plot(x, u[i], label='t={}'.format(i*tau))

plt.xlabel('x')
plt.ylabel('u')
plt.legend()
plt.title('Решение уравнения Хопфа методом Годунова')
plt.grid(True) 
plt.show()
```
